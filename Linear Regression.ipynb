```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd
# define path and set up dataframe
df_path = ('Linear_data.xlsx')
df1 = pd.read_excel(df_path,sheet_name='Train_Data')
df_path = ('Linear_data.xlsx')
df2 = pd.read_excel(df_path,sheet_name='Test_Data')
```

```python
# convert categorical variables Divisions to dummy variables
df1 = pd.get_dummies(df1, columns=['DIVISION'],drop_first=True)
#df1 = pd.get_dummies(df1, columns=['TOPNB30'], drop_first=True, dummy_na=False)
# Convert all columns to float data type
```

```python
# convert categorical variables Divisions to dummy variables
df2 = pd.get_dummies(df2, columns=['DIVISION'],drop_first=True)
# Convert all columns to float data type
```

```python
#Split the data into X test and y test
y_train = df1['B_E_CRIME']
X_train = df1.drop(columns=['B_E_CRIME','OCC_YEAR'])
#copying X into a variable called X_scaled for scaling to be used later
X_train_Scale = X_train.copy()
#scaling the data, even though we dont need for decision tree it will be useful when we do KNN
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Fit and transform the data
X_train_Scale = pd.DataFrame(scaler.fit_transform(X_train_Scale), columns=X_train_Scale.columns)
```

```python
#Split the data into X train and y train
y_test = df2['B_E_CRIME']
X_test = df2.drop(columns=['B_E_CRIME','OCC_YEAR'])
#copying X into a variable called X_scaled for scaling to be used later
X_test_Scale = X_test.copy()
#scaling the data, even though we dont need for decision tree it will be useful when we do KNN
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Fit and transform the data
X_test_Scale = pd.DataFrame(scaler.fit_transform(X_test_Scale), columns=X_test_Scale.columns)
```

```python
# Split the data into train and test sets
# y_test    X_test_Scale         y_train         X_train_Scale       
```

```python
# First, let's try forward stepwise regression
from stepwise_regression import step_reg

# Perform forward regression on the train set
forward_selected_predictors = step_reg.forward_regression(X_train_Scale, y_train, 0.05, verbose=True)

# Select the predictors from the train and test sets
X_train_Scale_forward = X_train_Scale[forward_selected_predictors]


# Fit the forward regression model
forward_model = sm.OLS(y_train, sm.add_constant(X_train_Scale_forward)).fit()
```

```python
# Print the summary of the forward regression model
print(forward_model.summary())
```

```python
# Second, let's try forward stepwise regression for test set
from stepwise_regression import step_reg

# Perform forward regression on the train set
forward_selected_predictors = step_reg.forward_regression(X_test_Scale, y_test, 0.05, verbose=True)

# Select the predictors from the train and test sets
X_test_Scale_forward = X_test_Scale[forward_selected_predictors]


# Fit the forward regression model
forward_model2 = sm.OLS(y_test, sm.add_constant(X_test_Scale_forward)).fit()
```

```python
# Print the summary of the forward regression model for test set
print(forward_model2.summary())
```

```python
# Predict on the train and test sets using the forward regression model
y_pred_forward_train = forward_model.predict(sm.add_constant(X_train_Scale_forward))
y_pred_forward_test = forward_model2.predict(sm.add_constant(X_test_Scale_forward))

# Calculate the MAE and RMSE for the forward regression model on the train and test sets
rmse_forward_train = mean_squared_error(y_train, y_pred_forward_train, squared=False)
rmse_forward_test = mean_squared_error(y_test, y_pred_forward_test, squared=False)
mae_forward_train = mean_absolute_error(y_train, y_pred_forward_train)
mae_forward_test = mean_absolute_error(y_test, y_pred_forward_test)

# Print the MAE and RMSE for the forward regression model on the train and test sets
print("RMSE (Forward Regression) - Train Set:", rmse_forward_train)
print("RMSE (Forward Regression) - Test Set:", rmse_forward_test)
print("MAE (Forward Regression) - Train Set:", mae_forward_train)
print("MAE (Forward Regression) - Test Set:", mae_forward_test)
```

```python
# Perform backward regression on the train set
backward_selected_predictors = step_reg.backward_regression(X_train_Scale, y_train, 0.05, verbose=True)

# Select the predictors from the train and test sets
X_train_Scale_backward = X_train_Scale[backward_selected_predictors]
X_test_Scale_backward = X_test_Scale[backward_selected_predictors]

# Fit the backward regression model
backward_model = sm.OLS(y_train, sm.add_constant(X_train_Scale_backward)).fit()
```

```python
# Print the summary of the backward regression model
print(backward_model.summary())
```

```python
# Perform backward regression on the test set
backward_selected_predictors = step_reg.backward_regression(X_test_Scale, y_test, 0.05, verbose=True)

# Select the predictors from the train and test sets
X_test_Scale_backward = X_test_Scale[backward_selected_predictors]


# Fit the backward regression model
backward_model2 = sm.OLS(y_test, sm.add_constant(X_test_Scale_backward)).fit()
```

```python
# Print the summary of the backward regression model on test set
print(backward_model2.summary())
```

```python
# Predict on the train and test sets using the backward regression model
y_pred_backward_train = backward_model.predict(sm.add_constant(X_train_Scale_backward))
y_pred_backward_test = backward_model2.predict(sm.add_constant(X_test_Scale_backward))

# Calculate the MAE and RMSE for the backward regression model on the train and test sets
mae_backward_train = mean_absolute_error(y_train, y_pred_backward_train)
mae_backward_test = mean_absolute_error(y_test, y_pred_backward_test)
rmse_backward_train = mean_squared_error(y_train, y_pred_backward_train, squared=False)
rmse_backward_test = mean_squared_error(y_test, y_pred_backward_test, squared=False)

# Print the MAE and RMSE for the backward regression model on the train and test sets
print("MAE (Backward Regression) - Train Set:", mae_backward_train)
print("MAE (Backward Regression) - Test Set:", mae_backward_test)
print("RMSE (Backward Regression) - Train Set:", rmse_backward_train)
print("RMSE (Backward Regression) - Test Set:", rmse_backward_test)
```

```python
# ElasticNet Regression
elasticnet_model2 = ElasticNetCV(alphas=np.logspace(-6, 6, 20), l1_ratio=np.linspace(0.1, 1.0, 10), cv=5)
elasticnet_model2.fit(X_test_Scale, y_test)

# Print ElasticNet Regression summary
print("ElasticNet Regression:")
print("Best alpha:", elasticnet_model2.alpha_)
print("Best L1 ratio:", elasticnet_model2.l1_ratio_)
print("Coefficients:", elasticnet_model2.coef_)
print("Intercept:", elasticnet_model2.intercept_)
print("")
```
