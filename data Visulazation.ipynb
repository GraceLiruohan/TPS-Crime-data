```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataframe from a CSV file or any other source
df = pd.read_csv('Break_and_Enter_Open_Data.csv')

# Select the dependent variable
dependent_variable = 'dependent_variable_name'

# Select the independent variables
independent_variables = ['independent_variable1', 'independent_variable2', 'independent_variable3']

# Calculate the correlation coefficient for each independent variable
correlations = {}
for independent_variable in independent_variables:
    correlation_coefficient = df[independent_variable].corr(df[dependent_variable])
    correlations[independent_variable] = correlation_coefficient

# Plot the correlation for each independent variable
for independent_variable, correlation_coefficient in correlations.items():
    plt.scatter(df[independent_variable], df[dependent_variable])
    plt.title('Correlation between {} and {}'.format(independent_variable, dependent_variable))
    plt.xlabel(independent_variable)
    plt.ylabel(dependent_variable)
    plt.text(df[independent_variable].min(), df[dependent_variable].max(), 'Correlation: {:.2f}'.format(correlation_coefficient), ha='left')
    plt.show()
```
